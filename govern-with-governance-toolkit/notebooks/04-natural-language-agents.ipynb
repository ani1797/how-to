# 04 — Natural Language Agents

Use Microsoft Agent Framework to drive the agents with natural language.

**Prerequisites**: Set `FOUNDRY_PROJECT_ENDPOINT` in `.env`. Without it, agents fall back to listing tools.

```python
import sys; sys.path.insert(0, '..')
from src.config import get_settings
settings = get_settings()
print('LLM configured:', settings.has_llm())
if not settings.has_llm():
    print('No LLM configured — running in no-LLM fallback mode')
```

```python
from src.agents.customer_service import CustomerServiceAgent
from src.models import AgentRole

cs = CustomerServiceAgent(role=AgentRole.CUSTOMER_SERVICE)
result = cs.ask('What is the loyalty tier for customer CUST-001?')
print(result)
```

```python
from src.agents.order_processing import OrderProcessingAgent

op = OrderProcessingAgent(role=AgentRole.STORE_MANAGER)
result = op.ask('Apply a 10% discount to SKU-300 Bluetooth Speaker')
print(result)
```

```python
from src.agents.returns import ReturnsAgent

ret = ReturnsAgent(role=AgentRole.CASHIER)
result = ret.ask('Can I return order ORD-002?')
print(result)
```
