# 06 — Deploy to Azure AI Foundry

Deploy governance-enforced agents to Azure AI Foundry using the Microsoft Agent Framework.

```python
import sys; sys.path.insert(0, '..')
from src.config import get_settings
settings = get_settings()
print('Foundry endpoint:', settings.foundry_project_endpoint or '(not configured)')
print('Model:', settings.model_deployment_name)
print('Policy dir:', settings.governance_policy_dir)
print('Audit log:', settings.governance_audit_log_path)
```

## Prerequisites

1. `az login` (Entra ID auth)
2. Set `FOUNDRY_PROJECT_ENDPOINT` in `.env`
3. Model deployment named `gpt-4o` (or update `MODEL_DEPLOYMENT_NAME`)

```python
# Run the deployment script
import subprocess, sys
result = subprocess.run([sys.executable, 'scripts/deploy_to_foundry.py'], capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print('ERROR:', result.stderr)
```

```python
# Validate post-deployment
result = subprocess.run([sys.executable, 'scripts/validate_deployment.py'], capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print('FAILURES:', result.stderr)
```

## What Runs in Foundry

The agents use `agent.ask(query)` which calls MAF's `Agent.run()` loop.
Governance enforcement is **in-process** inside the Foundry container — no sidecar, no network call.
